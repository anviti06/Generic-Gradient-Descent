```python
import numpy as np
train = np.loadtxt("train.csv", delimiter =",")
test = np.loadtxt("test.csv", delimiter =",")
```

```python
train_X = train[:,[0,1,2,3]]
train_Y = train[:,[4]]
train_x = np.array(train_X)
train_y = np.array(train_Y)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test.shape)
```

```python
b1 = np.ones((7176,1) )
b2 = np.ones((2392,1))
train_x = np.append(train_x, b1 ,axis = 1)
test = test[:,[0,1,2,3]]
test = np.append(test, b2 ,axis = 1)
```

```python
print(test.shape)
print(train.shape)
```

```python
def cost(x, y, m):
    N = len(x)
    cst=0
    for i in range(N):
        cst+=(1/N) * (( y[i] - (x[i] * m).sum())**2 )
    return cst
```

```python
def step_gd(x, y, learning_rate, m):
    N = len(x)
    slope_m =  np.zeros((1,5))
    new_m = np.zeros((1,5))
    for i in range(N):
        slope_m += (-2/N) * ( y[i] - (m * x[i]).sum() ) * x[i]
    new_m = m - learning_rate*slope_m
    return new_m
```

```python
def gd(x, y, learning_rate, num_iterations):
    m = np.zeros((1,5))
    for i in range(num_iterations):
        m = step_gd(x, y, learning_rate, m)
        print(i, " Cost : ",cost(x, y, m))
    return m
```

```python
def final(test_x , m):
    test_y = np.zeros((2392,1))
    for i in range(len(test_x)):
        test_y[i] += (m * test_x[i]).sum()
    return test_y
```

```python
def run():
    learning_rate = 0.00000096
    num_iterations = 8000
    m = gd(train_x, train_y, learning_rate, num_iterations)
    #print("M: ",m )
    return m
    
```

```python
m = run()
```

```python
test_y = final(test, m )
```
